# Dataset before preprocessing

```
import os
import struct
import matplotlib.pyplot as plt
import pandas as pd

file_size = os.path.getsize('drive/MyDrive/UmpireAI/data/data_dump_03_30_21.bin')
data_points_total_count = file_size // 4
data_label_points = 1
data_points_per_vibration = 100 + data_label_points

file = open('drive/MyDrive/UmpireAI/data/data_dump_03_30_21.bin', 'rb')
numbers = list(struct.unpack('f' * data_points_total_count, file.read(4 * data_points_total_count)))
file.close()

with open('drive/MyDrive/UmpireAI/data/readable.csv', 'w') as f:
  for index, item in enumerate(numbers):
    f.write("%s , "  %( item))
    if (index != 0 and (index + 1) % data_points_per_vibration == 0):
      f.write("\n")
      
    
    


plt.plot(numbers)
plt.show()








```
